```python
import pandas as pd

# Load dataset
file_path = "/content/most_watched_anime_dataset_100_entries.csv"
anime_df = pd.read_csv(file_path)

# Fill missing values
anime_df["Anime Name"].fillna("Unknown", inplace=True)
anime_df["Most Watched in Country"].fillna("Unknown", inplace=True)
anime_df["Genre"].fillna("Unknown", inplace=True)
anime_df["Number of Episodes"].fillna(anime_df["Number of Episodes"].median(), inplace=True)
anime_df["Duration per Episode (minutes)"].fillna(anime_df["Duration per Episode (minutes)"].median(), inplace=True)

# Convert Release Year to integer
anime_df["Release Year"] = anime_df["Release Year"].astype("Int64")

# Drop duplicates
anime_df.drop_duplicates(inplace=True)

# Display dataset info
print(anime_df.info())
print(anime_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("darkgrid")

# 1. Top 10 Most-Watched Anime
most_watched = anime_df['Anime Name'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=most_watched.values, y=most_watched.index, palette='viridis')
plt.xlabel("Count")
plt.ylabel("Anime Name")
plt.title("Top 10 Most-Watched Anime")
plt.show()

# 2. Genre Distribution
genre_counts = anime_df['Genre'].value_counts()
plt.figure(figsize=(12,6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='magma')
plt.xlabel("Count")
plt.ylabel("Genre")
plt.title("Anime Genre Distribution")
plt.show()

# 3. Top 10 Highest-Rated Anime
top_rated = anime_df.sort_values(by='Ratings', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_rated['Ratings'], y=top_rated['Anime Name'], palette='coolwarm')
plt.xlabel("Ratings")
plt.ylabel("Anime Name")
plt.title("Top 10 Highest-Rated Anime")
plt.show()

# 4. Budget vs. Ratings
plt.figure(figsize=(10,6))
sns.scatterplot(data=anime_df, x='Budget (in Million USD)', y='Ratings', hue='Genre', palette='tab10', alpha=0.8)
plt.xlabel("Budget (in Million USD)")
plt.ylabel("Ratings")
plt.title("Budget vs. Ratings")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 5. Popular Animation Studios
studio_counts = anime_df['Animation Studio Name'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=studio_counts.values, y=studio_counts.index, palette='rocket')
plt.xlabel("Number of Anime")
plt.ylabel("Animation Studio")
plt.title("Top 10 Most Popular Animation Studios")
plt.show()



```
